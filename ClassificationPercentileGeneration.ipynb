```python
# COLAB SCRIPT FOR PROCESSING DATASET AND CALCULATING PERCENTILES
from ultralytics import YOLO
import cv2
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import json

# Load your trained model
model = YOLO('/content/gdrive/MyDrive/microplastics_augmented/CV Project/best.pt')

# Make sure to use GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Path to your dataset
dataset_path = '/content/gdrive/MyDrive/microplastics_augmented/train/images'  # Update this path
image_paths = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path)
               if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Lists to store metrics
all_metrics = []

# Process each image and extract metrics
for img_path in tqdm(image_paths, desc="Processing images"):
    # Load image
    img = cv2.imread(img_path)

    # Run detection
    results = model(img)

    # Extract boxes from results
    boxes = results[0].boxes

    if len(boxes) > 0:
        # Number of plastics
        num_plastics = len(boxes)

        # Calculate sizes (based on bounding box area in pixels)
        # boxes.xyxy is in format [x1, y1, x2, y2]
        if hasattr(boxes, 'xyxy') and boxes.xyxy is not None:
            xyxy = boxes.xyxy.cpu().numpy()
            widths = xyxy[:, 2] - xyxy[:, 0]
            heights = xyxy[:, 3] - xyxy[:, 1]
            areas = widths * heights
            avg_size = np.mean(areas)

            # Calculate aspect ratios (always >= 1)
            aspect_ratios = np.maximum(widths / heights, heights / widths)
            avg_aspect_ratio = np.mean(aspect_ratios)
        else:
            # Fallback if xyxy not available
            avg_size = 0
            avg_aspect_ratio = 0

        # Store metrics
        all_metrics.append({
            'image': os.path.basename(img_path),
            'num_plastics': num_plastics,
            'avg_size': avg_size,
            'avg_aspect_ratio': avg_aspect_ratio
        })
    else:
        # No detections
        all_metrics.append({
            'image': os.path.basename(img_path),
            'num_plastics': 0,
            'avg_size': 0,
            'avg_aspect_ratio': 0
        })

# Convert to DataFrame
metrics_df = pd.DataFrame(all_metrics)

# Calculate percentiles
percentiles = [25, 50, 75, 90]
percentile_values = {}

for metric in ['num_plastics', 'avg_size', 'avg_aspect_ratio']:
    percentile_values[metric] = {}
    for p in percentiles:
        percentile_values[metric][p] = float(metrics_df[metric].quantile(p/100))

# Print percentiles
print("Calculated Percentile Values:")
for metric, values in percentile_values.items():
    print(f"\n{metric}:")
    for p, val in values.items():
        print(f"  {p}th percentile: {val:.2f}")

# Visualize distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
metrics = ['num_plastics', 'avg_size', 'avg_aspect_ratio']
titles = ['Microplastic Count', 'Average Size (px²)', 'Average Aspect Ratio']

for i, (metric, title) in enumerate(zip(metrics, titles)):
    sns.histplot(metrics_df[metric], kde=True, ax=axes[i])
    axes[i].set_title(title)

    # Add percentile lines
    for p, color in zip(percentiles, ['blue', 'green', 'orange', 'red']):
        val = percentile_values[metric][p]
        axes[i].axvline(val, color=color, linestyle='--',
                      label=f'{p}th percentile: {val:.2f}')

    axes[i].legend()

plt.tight_layout()
plt.savefig('metric_distributions.png')
plt.show()

# Save the percentile values to a JSON file for later use in Streamlit
with open('percentile_values.json', 'w') as f:
    json.dump(percentile_values, f)

# Also save the full metrics dataframe for reference
metrics_df.to_csv('dataset_metrics.csv', index=False)
```
